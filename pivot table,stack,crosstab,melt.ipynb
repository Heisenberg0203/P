```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather_by_cities.csv",parse_dates=['day'])
df
```

```python
pdf=df.pivot_table(index='day',columns='city',aggfunc=np.mean)
pdf
```

```python
df.pivot_table(index=pd.Grouper(freq='172800s',key='day'),columns='city')
```

### To convert back we will be using stack method

```python
df=pdf.stack().reset_index()
df
```

```python
pd.crosstab(df.day,df.city[0:11],margins=True,normalize=True)
```
